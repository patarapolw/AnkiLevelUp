```python
from AnkiTools.tools.read import readAnki2
```

```python
anki = readAnki2('Chinese.anki2')
```

```python
anki.loadQuery()
```

```python
def wrapHtml(inner, wrapper):
    head, tail = wrapper.split('></')
    return head + '>' + inner + '</' + tail
```

```python
old_length = 0
output = ''
for deck_name, deck_id in sorted((deck['name'], deck['did']) for deck in anki.decks):
    names = deck_name.split('::')
    new_length = len(names)
    content = names[-1]
    
    if new_length > old_length:
        while new_length > old_length:
            old_length += 1
            output += '<ul><li>'
        output += content
    elif new_length == old_length:
        output += wrapHtml(content, '<li></li>')
    elif new_length < old_length:
        while new_length < old_length:
            old_length -= 1
            output += '</li></ul>'
        output += content
    
    old_length = new_length
output += '</li></ul>'
output
```
